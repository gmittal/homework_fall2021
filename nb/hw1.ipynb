```python
import glob
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn')
plt.rcParams.update({'font.size': 22})
np.random.seed(seed=1)


def set_size(width_pt, fraction=1, subplots=(1, 1)):
    """Set figure dimensions to sit nicely in our document.
  
    Source: https://jwalton.info/Matplotlib-latex-PGF/
    """
    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)
```

### Problem 1.3
Behavioral cloning with HalfCheetah-v2 and number of training steps ablated.

```python
x = []
y = []
yerr = []
expert_y = []
for log in glob.glob("logs/*.log"):
    with open(log) as f:
        data = f.readlines()
    
    expert = None
    mu = None
    std = None
    for row in data:
        n = re.findall("\d+\.\d+", row)
        if row.find('Eval_AverageReturn') > -1:
            mu = float(n[0])
        if row.find('Eval_StdReturn') > -1:
            std = float(n[0])
        if row.find('Initial_DataCollection_AverageReturn') > -1:
            expert = float(n[0])
    
    iters = int(log.split('-')[-1].split('.')[0])
    x.append(iters)
    y.append(mu)
    yerr.append(std)
    expert_y.append(expert)

x = np.array(x)
y = np.array(y)
yerr = np.array(yerr)
expert_y = np.array(expert_y)

order = sorted(range(len(x)), key=lambda i: x[i])
x = x[order]
y = y[order]
yerr = yerr[order]
expert_y = expert_y[order]
    
# plot graph
fig, ax = plt.subplots(1, 1, figsize=(width, height))
line1, = ax.plot(x, y)
line2, = ax.plot(x, expert_y)
line1.set_label('BC Policy')
line2.set_label('Expert Policy')

ax.errorbar(x, y, yerr=yerr, fmt='o')
ax.set_ylabel('Policy average return on HalfCheetah-v2')
ax.set_xlabel('Number of training iterations')
# ax.set_xscale('log')
ax.legend(frameon=True, loc='lower right')
```

### Problem 2.2
DAgger agent on Ant-v2 and HalfCheetah-v2.

```python
results = {}
bc_results = {
    'Ant-v2': (4573.38818359375, 108.95249938964844),
    'HalfCheetah-v2': (3781.06591796875, 142.93663024902344),
}

for log in glob.glob("logs/dagger/*.log"):
    with open(log) as f:
        data = f.readlines()
        
    iteration = 0
    x = []
    y = []
    yerr = []
    expert_y = []
    for row in data:
        n = re.findall("\d+\.\d+", row)
        if row.find('Iteration') > -1:
            x.append(iteration)
            iteration += 1
        if row.find('Eval_AverageReturn') > -1:
            mu = float(n[0])
            y.append(mu)
        if row.find('Eval_StdReturn') > -1:
            std = float(n[0])
            yerr.append(std)
        if row.find('Initial_DataCollection_AverageReturn') > -1:
            expert = float(n[0])
            expert_y.append(expert)

    if 'ant' in log:
        results['Ant-v2'] = (x, y, yerr, expert_y)
    
    if 'hc' in log:
        results['HalfCheetah-v2'] = (x, y, yerr, expert_y)
        
fig, axes = plt.subplots(1, 2, figsize=(width * 2, height))

for i, name in enumerate(results.keys()):
    x, y, yerr, expert_y = results[name]
    line1, = axes[i].plot(x, y)
    line2, = axes[i].plot(x, expert_y)
    line3, = axes[i].plot(x, [bc_results[name][0]] * len(x))
    line1.set_label('DAgger')
    line2.set_label('Expert')
    line3.set_label('BC')
    axes[i].errorbar(x, y, yerr=yerr, fmt='o')
#     axes[i].errorbar(x, [bc_results[name][0]] * len(x), yerr=[bc_results[name][1]] * len(x), fmt='o')

    axes[i].set_title(name)
    axes[i].set_ylabel(f'Policy average return on {name}')
    axes[i].set_xlabel('Iterations')
    axes[i].set_xlim(-0.5, 10)
    axes[i].legend(frameon=True, loc='lower right')
```
