```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
plt.rcParams.update({'font.size': 22})
np.random.seed(seed=1)


def set_size(width_pt, fraction=1, subplots=(1, 1)):
    """Set figure dimensions to sit nicely in our document.
  
    Source: https://jwalton.info/Matplotlib-latex-PGF/
    """
    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)
```

```python
width, height = set_size(600 * 1.5, subplots=(1, 1))
```

### Experiment 2

```python
x_train = [0]
y_train = [-167.1]
x_eval = [0]
y_eval = [-45.99]

fig, ax = plt.subplots(1, 1, figsize=(width, height))
ax.scatter(x_train, y_train)
ax.scatter(x_eval, y_eval)
ax.set_ylabel('Average Return')
ax.set_xlabel('Iteration')
ax.set_title('Training and Evaluation Performance')
```

### Experiment 3

```python
runs = glob.glob('logs/p3/*-Eval_AverageReturn.csv')

fig, ax = plt.subplots(1, 1, figsize=(width, height))

for run in runs:
    df = pd.read_csv(run)
    
    std_run = ''.join(run.split('-Eval_AverageReturn.csv') + ['-Eval_StdReturn.csv'])
    std_df = pd.read_csv(std_run)
    std = std_df['Value'].values
    
    param_str = run.split('q3_')[-1].split('_')[0]
    steps = df['Step'].values
    values = df['Value'].values
    line, = ax.plot(steps, values)
    ax.fill_between(steps, values-std, values+std, alpha=0.3)
    line.set_label(param_str)
    
ax.set_ylabel('Average eval return')
ax.set_xlabel('Iteration')
ax.set_title('On-policy data collection with iterative training')
ax.legend(frameon=False, loc='lower right')
```
